##                                                          BETA DISTRIBUTION

```python
import numpy.random as r
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import integrate
import math
from scipy.stats import beta
```

```python
alpha=int(input("Enter the value of α:"))
beta_param=int(input("Enter the value of β:"))
```

Defining function B(α,β)

```python
def beta_function(alpha,beta_param):
    if alpha<=0 or beta_param<=0:
        raise ValueError("Alpha and Beta must be positive real numbers")
    
    def integrand(x):
        return (x**(alpha-1))*((1-x)**(beta_param-1))
    result, _ = integrate.quad(integrand,0,1)
    return result
```

```python
print(f"B({alpha},{beta_param})=",beta_function(alpha,beta_param))
```

-------

#### Expected value , Mean and Variance Calculation

```python
def expected_val_function(alpha,beta_param):
    return  (alpha)/(alpha+beta_param)
expected_val = expected_val_function(alpha,beta_param)

def variance_fun(alpha,beta_param):
    variance = (alpha*beta_param)/(((alpha+beta_param)**2)*(alpha+beta_param+1))
    return variance

def mode_function(alpha, beta_param):
    if alpha > 1 and beta_param > 1:
        return (alpha - 1) / (alpha + beta_param - 2)
    elif alpha <= 1 or beta_param <= 1:
        return None
mode_val = mode_function(alpha, beta_param)

print(f"E(x)=",expected_val)
print(f"Mean=",expected_val)
print(f"Var(x)=",variance_fun(alpha,beta_param))
print(f"Mode=",mode_val)
```
