## EXPONENTIAL DISTRIBUTION

alpha = 1

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import integrate
from scipy.stats import expon
import math
```

```python
x = float(input("Enter the value of x:"))
beta_param=float(input("Enter the value of β:"))
```

```python
alpha = 1  
def exponential_func_pdf(x,beta_param):
    if beta_param<=0:
        raise ValueError("Beta must be positive real number")
    if x<0:
        return 0.0

    ex_coeff = (beta_param)*math.exp(-beta_param*x)
    return ex_coeff
pdf_val = exponential_func_pdf(x,beta_param)
```

```python
print(f"pdf of f({x},{beta_param})=",exponential_func_pdf(x,beta_param))
```

```python
def expected_val_function(beta_param):
    return  (1)/(beta_param)
expected_val = expected_val_function(beta_param)

def variance_fun(beta_param):
    variance = 1/(beta_param**2)
    return variance

print(f"E(x)=",expected_val)
print(f"Mean=",expected_val)
print(f"Var(x)=",variance_fun(beta_param))
```

### PLOT

```python
x_values = np.linspace(0,1, 500)
```

```python
pdf_values = [exponential_func_pdf(x,beta_param) for x in x_values]
```

```python
sns.set(style="whitegrid")
plt.plot(x_values,pdf_values,label=f'Exponential Distribution (β={beta_param})')

plt.title(f"Exponential Distribution(β={beta_param})",fontsize=15)
#fill
plt.fill_between(x_values, pdf_values, alpha=0.3, color="skyblue")
#mean
plt.axvline(expected_val, color="red", linestyle="--", label=f'Mean (E(x)) = {expected_val:.2f}')
#labels
plt.xlabel('x',fontsize=12)
plt.ylabel('Probability Density',fontsize=12)
#95th percentile
# 95th percentile
percentile_95 = -math.log(1 - 0.95) / beta_param
plt.axvline(percentile_95, color="green", linestyle="--", label=f'95th percentile ≈ {percentile_95:.2f}')
plt.legend()
plt.show()
```
