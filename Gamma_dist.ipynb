### GAMMA DISTRIBUTION PDF

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import math 
from scipy import integrate
from scipy.stats import gamma
```

```python
x = float(input("Enter the value of x:"))
alpha=float(input("Enter the value of α:"))
beta_param=float(input("Enter the value of β:"))
```

Defining function γ(α,β)
alpha controls the shape and beta controls the scale

```python
def gamma_function_pdf(x,alpha,beta_param):
    if alpha<=0 or beta_param<=0:
        raise ValueError("Alpha and Beta must be positive real numbers")
    if x<0:
        return 0.0
    coefficient = (beta_param**alpha)/math.gamma(alpha)
    prob_dens_fn = coefficient * (x ** (alpha - 1)) * math.exp(-beta_param* x)
    return prob_dens_fn
```

```python
print(f"pdf of f({x},{alpha},{beta_param})=",gamma_function_pdf(x,alpha,beta_param))
```

#### Expected value , Mean and Variance Calculation

```python
def expected_val_function(alpha,beta_param):
    return  (alpha)/(beta_param)
expected_val = expected_val_function(alpha,beta_param)

def variance_fun(alpha,beta_param):
    variance = (alpha)/(beta_param**2)
    return variance

print(f"E(x)=",expected_val)
print(f"Mean=",expected_val)
print(f"Var(x)=",variance_fun(alpha,beta_param))
```

-------

### PLOT

```python
x_values = np.linspace(0,10,1000)
```

```python
pdf_values = [gamma_function_pdf(x,alpha,beta_param) for x in x_values]
```

```python
sns.set(style="whitegrid")
plt.plot(x_values,pdf_values,label=f'Gamma Distribution (α={alpha},β={beta_param})')

plt.title(f"Gamma Distribution(α={alpha},β={beta_param})",fontsize=15)
#fill
plt.fill_between(x_values, pdf_values, alpha=0.3, color="skyblue")
#mean
plt.axvline(expected_val, color="red", linestyle="--", label=f'Mean (E(x)) = {expected_val:.2f}')
#labels
plt.xlabel('x',fontsize=12)
plt.ylabel('Probability Density',fontsize=12)

plt.legend()
plt.show()
```
